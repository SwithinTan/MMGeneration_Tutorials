## Python API 方式调用

### 导入工具包

````python
# 导入mmcv和mmgeneration
import mmcv
from mmgen.apis import init_model, sample_unconditional_model

# 导入numpy和matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

### 载入 StyleGAN V2 预训练模型

注意，`config_file`和`checkpoint_file`需要对应。

````python
# 教堂图像生成模型
# 指定config文件路径
config_file = 'configs/styleganv2/stylegan2_c2_lsun-church_256_b4x8_800k.py'

# 指定预训练模型权重文件路径
checkpoint_file = 'https://download.openmmlab.com/mmgen/stylegan2/official_weights/stylegan2-church-config-f-official_20210327_172657-1d42b7d1.pth'

# 输出图像宽高像素尺寸
img_size = 256
````

````python
# # 人脸图像生成模型
# # 指定config文件路径
# config_file = 'configs/styleganv2/stylegan2_c2_ffhq_1024_b4x8.py'

# # 指定预训练模型权重文件路径
# checkpoint_file = 'https://download.openmmlab.com/mmgen/stylegan2/stylegan2_c2_ffhq_1024_b4x8_20210407_150045-618c9024.pth'

# # 输出图像宽高像素尺寸
# img_size = 1024
````

### 初始化模型

````python
# 初始化generatvie
model = init_model(config_file, checkpoint_file, device='cuda:0')
````

### 生成若干张图像

````python
# 生成 n的平方 张图像
n = 8
````

````python
# 生成图像
fake_imgs = sample_unconditional_model(model, n * n)
````

````python
fake_imgs.shape
````

### 展示单张生成的图像

````python
# 将torch张量转为numpy的array
fake_imgs = fake_imgs.numpy()
````

````python
fake_imgs.shape
````

````python
# 选择要展示的图片索引号
index = 41

# 分别抽取RGB三通道图像，归一化为0-255的uint8自然图像
RGB = np.zeros((img_size, img_size, 3))
RGB[:,:,0] = fake_imgs[index][2]
RGB[:,:,1] = fake_imgs[index][1]
RGB[:,:,2] = fake_imgs[index][0]

RGB = 255 * (RGB - RGB.min()) / (RGB.max()-RGB.min())
RGB = RGB.astype('uint8')
plt.imshow(RGB)
plt.show()
````

### n行n列展示生成的图像

````python
def show_fake_img_rgb(index):
    '''输入索引号，展示对应的图像'''

    # 分别抽取RGB三通道图像，归一化为0-255的uint8自然图像
    RGB = np.zeros((img_size,img_size,3))
    RGB[:,:,0] = fake_imgs[index][2]
    RGB[:,:,1] = fake_imgs[index][1]
    RGB[:,:,2] = fake_imgs[index][0]

    RGB = 255 * (RGB - RGB.min()) / (RGB.max()-RGB.min())
    RGB = RGB.astype('uint8')
    return RGB
#     plt.imshow(RGB)
#     plt.show()
````

````python
fig, axes = plt.subplots(nrows=n, ncols=n, sharex=True,figsize=(20,20))
for i in range(n*n):
    axes[i//n, i%n].imshow(show_fake_img_rgb(i))
    axes[i//n, i%n].axis('off') # 关闭坐标轴显示
fig.suptitle('mmgeneration demo',fontsize=50)
plt.show()
````
